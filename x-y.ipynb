```python
import numpy as np
import pandas as pd
A = np.zeros([2])
B = np.zeros([2])
C = np.zeros([2])
for i in range(2):
    A[i] = float(input(f'A{i+1} = '))
    B[i] = float(input(f'B{i+1} = '))
    C[i] = float(input(f'C{i+1} = '))
```

```python
P = float(input('Vlue of pressure in mmHg : '))
```

```python
saturated_T = np.array([])
for i in range(2):
    TS = (B[i]/(A[i]-np.log10(P))-C[i])
    saturated_T = np.append(saturated_T,TS)
T = np.arange(saturated_T[0],saturated_T[1],1)
```

```python
saturated_T = np.array([])
for i in range(2):
    TS = (B[i]/(A[i]-np.log10(P))-C[i])
    saturated_T = np.append(saturated_T,TS)
T = np.arange(saturated_T[0],saturated_T[1],1)
PA = np.array([])
for i in range(len(T)):
    P = 10**(A[0]-(B[0]/(T[i]+C[0])))
    PA = np.append(PA,P)
PB  = np.array([])
for i in range(len(T)):
    P = 10**(A[1]-(B[1]/(T[i]+C[1])))
    PB = np.append(PB,P)
x1 = np.array([])
for i in range(len(T)):
    x = (P-PB[i])/(PA[i]-PB[i])
    x1 = np.append(x1,x)
y1 = np.array([])
for i in range(len(T)):
    y = (PA[i]*x1[i])/P
    y1 = np.append(y1,y)
alphas = np.array([])
for i in range(len(T)):
    alpha = PA[i]/PB[i]
    alphas = np.append(alphas,alpha)
table = pd.DataFrame(T,columns=['Temperature'])
table['PA'] = PA
table['PB'] = PB
table['x1'] = x1
table['y1'] = y1
table['alpha'] = alphas
table
```
